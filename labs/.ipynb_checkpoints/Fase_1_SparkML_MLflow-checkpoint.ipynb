# Fase 1: Spark ML + MLflow Baseline
Este notebook establece el baseline del proyecto utilizando **Logistic Regression** sobre variables numéricas, integrando la ingesta desde **BigQuery** y el tracking con **MLflow**.

```python
import os
from pyspark.sql import SparkSession
import mlflow
import mlflow.spark
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

## 1. Configuración de SparkSession y BigQuery
Configuramos Spark para usar el conector de BigQuery y autenticarnos con la Service Account.

```python
# Ruta al secreto montado en el contenedor
GCP_KEY_PATH = "/opt/spark/gcp-key.json"
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = GCP_KEY_PATH

spark = SparkSession.builder \
    .appName("Fase1_SparkML_Baseline") \
    .config("spark.jars.packages", "com.google.cloud.spark:spark-bigquery-with-dependencies_2.12:0.35.1") \
    .config("spark.hadoop.google.cloud.auth.service.account.enable", "true") \
    .config("spark.hadoop.google.cloud.auth.service.account.json.keyfile", GCP_KEY_PATH) \
    .getOrCreate()

print("Spark Session iniciada con soporte para BigQuery")
```

## 2. Ingesta de Datos desde BigQuery
Leemos el dataset `usa_names.usa_1910_2013`.

```python
table = "bigquery-public-data.usa_names.usa_1910_2013"

df = spark.read.format("bigquery") \
    .option("table", table) \
    .load()

df.printSchema()
df.show(5)
```

## 3. Preparación de Datos y Pipeline de ML
Seleccionamos variables numéricas para el baseline.

```python
# 1. Selección de columnas reales del dataset de nombres
# Columnas: name, gender, state, year, number
# Queremos predecir 'gender' (nuestro label) usando las demás.

# Indexamos las columnas categóricas
state_indexer = StringIndexer(inputCol="state", outputCol="state_index") \
    .setHandleInvalid("keep")  # <---  
name_indexer = StringIndexer(inputCol="name", outputCol="name_index") \
    .setHandleInvalid("keep")  # <--- 

# El target (lo que queremos predecir) es el género
label_indexer = StringIndexer(inputCol="gender", outputCol="label") \
    .setHandleInvalid("keep")

# 2. Configuración de Features
# Usaremos el nombre (indexado), el estado (indexado), el año y la cantidad
feature_cols = ["name_index", "state_index", "year", "number"]
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")

# 3. Definición del Modelo
lr = LogisticRegression(featuresCol="features", labelCol="label", maxIter=10)

# 4. Construcción del Pipeline
pipeline = Pipeline(stages=[state_indexer, name_indexer, label_indexer, assembler, lr])

# 5. Split de datos y entrenamiento
train_df, test_df = df.randomSplit([0.8, 0.2], seed=42)

# Ajustar el pipeline
model = pipeline.fit(train_df)
predictions = model.transform(test_df)

predictions.select("name", "gender", "label", "prediction", "probability").show(5)
```

## 4. Entrenamiento y Tracking con MLflow
Registramos el experimento en el servidor local de MLflow.

```python
import mlflow
import mlflow.spark
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

mlflow.set_tracking_uri("http://mlflow:5000")
mlflow.set_experiment("Spark_ML_Baseline")

numeric_features = ["name_index", "state_index", "year", "number"]

with mlflow.start_run(run_name="LogisticRegression_Numerical"):
    # 1. Log de parámetros
    mlflow.log_param("maxIter", 10)
    mlflow.log_param("features", numeric_features)

    # 2. Fit model
    model = pipeline.fit(train_df)

    # 3. Predicciones
    predictions = model.transform(test_df)

    # 4. Evaluación (Multiclase para evitar errores de índices nuevos)
    evaluator = MulticlassClassificationEvaluator(
        labelCol="label",
        predictionCol="prediction",
        metricName="accuracy"
    )

    

    # Calculamos Accuracy
    accuracy = evaluator.evaluate(predictions)
    mlflow.log_metric("accuracy", accuracy)

    # 5. Opcional: Calcular F1-Score (muy útil si los datos están desbalanceados)
    f1_score = evaluator.evaluate(predictions, {evaluator.metricName: "f1"})
    mlflow.log_metric("f1_score", f1_score)

    print(f"Model Accuracy: {accuracy:.4f}")
    print(f"Model F1-Score: {f1_score:.4f}")

    # 6. Log del modelo (Pipeline completo)
    mlflow.spark.log_model(model, "model")

    # Añadimos un Tag para saber qué dataset usamos
    mlflow.set_tag("dataset", "usa_names_1910_2013")

    print("Experimento registrado con éxito en MLflow")
```

## 5. Inferencia semántica
Registramos el experimento en el servidor local de MLflow.

```python
from pyspark.sql.functions import pandas_udf
import pandas as pd
from transformers import pipeline

# 1. El modelo se define y se prepara para Broadcast
MODEL_ID = "distilbert-base-uncased-finetuned-sst-2-english"
model_bc = spark.sparkContext.broadcast(MODEL_ID)

# 2. La Pandas UDF que utiliza la potencia de Arrow definida en tu conf
@pandas_udf("string")
def llm_sentiment_analysis(titles: pd.Series) -> pd.Series:
    # La pipeline se carga en el worker usando el nombre distribuido
    pipe = pipeline("sentiment-analysis", model=model_bc.value, device=-1)

    # Inferencia vectorizada gracias a Arrow
    predictions = pipe(titles.tolist(), truncation=True)
    return pd.Series([p['label'] for p in predictions])

# 3. Transformación del DataFrame original de la Fase 1
df_final = df.withColumn("sentiment_label", llm_sentiment_analysis(df["name"]))

# 4. Registro del experimento en MLflow
with mlflow.start_run(run_name="Fase_2_LLM_Inference"):
    df_final.select("name", "sentiment_label").show(5)
    mlflow.log_param("arrow_enabled", "true")
    mlflow.log_param("model_used", MODEL_ID)
```

```python

# 1. Configuración del nuevo experimento en MLflow
mlflow.set_experiment("USA_Names_Culture_Analysis")

# 2. Definición del modelo de Lenguaje (Raíz Lingüística)
# Este modelo es mucho más coherente para analizar nombres que uno de sentimiento.
MODEL_CULTURE = "papluca/xlm-roberta-base-language-detection"
model_bc_culture = spark.sparkContext.broadcast(MODEL_CULTURE)

# 3. Pandas UDF optimizada con Apache Arrow (ya configurado en tu spark-defaults.conf)
@pandas_udf("string")
def detect_culture_udf(names: pd.Series) -> pd.Series:
    # Inferencia distribuida en los workers
    classifier = pipeline("text-classification", model=model_bc_culture.value, device=-1)
    results = classifier(names.tolist(), truncation=True)
    return pd.Series([res['label'] for res in results])

# 4. Ejecución y Registro del Experimento
with mlflow.start_run(run_name="Culture_Inference_Detection"):
    
    # Aplicamos la IA sobre la columna 'name' del DataFrame 'df' original
    df_culture = df.withColumn("linguistic_origin", detect_culture_udf(df["name"]))
    
    # Mostramos los resultados para validar la inferencia
    print("Muestra de nombres con su raíz lingüística detectada:")
    df_culture.select("name", "gender", "linguistic_origin").show(10)
    
    # Registro de metadatos (Params)[cite: 2]
    mlflow.log_param("model_name", MODEL_CULTURE)
    mlflow.log_param("data_source", "bigquery_usa_names")
    
    # --- GENERACIÓN DE ARTEFACTO (Resultado de Negocio) ---
    # Guardamos una muestra de los resultados en un CSV para el equipo de Marketing
    result_sample = df_culture.select("name", "gender", "linguistic_origin").limit(100).toPandas()
    result_sample.to_csv("name_culture_report.csv", index=False)
    
    # Subimos el archivo a MLflow como un Artefacto[cite: 2]
    mlflow.log_artifact("name_culture_report.csv")
    
    print("¡Experimento finalizado! El reporte CSV se ha guardado como un artefacto en MLflow.")
```
